````python
import numpy as np
from matplotlib import pyplot as plt
````

```np.plot()``` function is used to plot a line graph.
Similarly<br> ```np.bar()``` function for bar chart,<br>```np.hist()``` funvtion for histogram,<br>```np.pie()``` function for pie chart <br>
```np.scatter()``` function is used to print scatter plot graph etc

````python
x =np.array([0,0,2])
y = np.array([6,250,40])
plt.plot(x,y)
plt.legend("blue")
plt.xlabel("height")
plt.ylabel("weight")
plt.show()
````

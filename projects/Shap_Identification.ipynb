## Shape identification Model
---

### Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

### Sample dataset (x, y, label)
---

```python
data = np.array([
    [1, 2, 0],  # Square
    [2, 2, 0],  # Square
    [2, 3, 0],  # Square
    [3, 3, 0],  # Square
    [6, 5, 1],  # Triangle
    [7, 5, 1],  # Triangle
    [7, 4, 1],  # Triangle
    [6, 4, 1],  # Triangle
    [4, 6, 2],  # Circle
    [8, 6, 2],  # Circle
    [5, 8, 2],  # Circle
])
```

### Split the data into features (X) and labels (y)
---

```python
X = data[:, :-1]  # Features (x, y coordinates)
y = data[:, -1]    # Labels (shape)

# Create a KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
```

### Generate random data points for visualization
---

```python
np.random.seed(0)
n_samples = 50
random_data = np.random.rand(n_samples, 2) * 10

predicted_labels = knn.predict(random_data)
```

### Separate the data points by shape
---

```python
squares = random_data[predicted_labels == 0]
triangles = random_data[predicted_labels == 1]
circles = random_data[predicted_labels == 2]
```

### Plot the shapes using Matplotlib
---

```python
plt.scatter(squares[:, 0], squares[:, 1], color='blue', marker='s', label='Square')
plt.scatter(triangles[:, 0], triangles[:, 1], c='red', marker='^', label='Triangle')
plt.scatter(circles[:, 0], circles[:, 1], c='green', marker='o', label='Circle')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Shapes recognition')
plt.show()
```

### Test case

```python
np.random.seed(0)
n_test_samples = 20
random_test_data = np.random.rand(n_test_samples, 2) * 10
predicted_labels = knn.predict(random_test_data)

for i in range(len(predicted_labels)):
    if predicted_labels[i] == 0:
        plt.scatter(random_test_data[i, 0], random_test_data[i, 1], c='blue', marker='s', label='Predicted Square')
    elif predicted_labels[i] == 1:
        plt.scatter(random_test_data[i, 0], random_test_data[i, 1], c='red', marker='^', label='Predicted Triangle')
    elif predicted_labels[i] == 2:
        plt.scatter(random_test_data[i, 0], random_test_data[i, 1], c='green', marker='o', label='Predicted Circle')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Shapes Recognition for Unlabeled Data')

plt.show()
```
